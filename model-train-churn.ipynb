```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('df.csv')
```

```python
df.head()
```

```python
df = df.drop('Unnamed: 0',axis=1)
```

```python
df
```

```python
x = df.drop('Churn Value',axis=1)
```

```python
print(x)
```

```python
y = df['Churn Value']
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# DecisionTreeClassifier

```python
model_dt = DecisionTreeClassifier(criterion='gini',random_state=100,max_depth=7,min_samples_leaf=8)
```

```python
model_dt.fit(x_train,y_train)
```

```python
y_pred = model_dt.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(accuracy)
```

since we can ses that our accuracy is less beaucse data is imbalencd so we need to make it balenced by methods like upsampling or down sampling

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

so we can see its biased model so we need to do upsampling to improve our model to do actual prediction

```python
print("Missing values before fix:")
print(x.isnull().sum()[x.isnull().sum() > 0])
x = x.fillna(0)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

# makng data balnced using upsampling

```python
sm = SMOTEENN()
X_resampled1, y_resampled1 = sm.fit_resample(x,y)
```

```python
xr_train1,xr_test1,yr_train1,yr_test1=train_test_split(X_resampled1, y_resampled1,test_size=0.2)
```

```python
model_rf_smote=RandomForestClassifier(n_estimators=1000, criterion='gini', random_state = 100,max_depth=6, min_samples_leaf=8)
```

```python
model_rf_smote.fit(xr_train1,yr_train1)
```

```python
yr_predict1 = model_rf_smote.predict(xr_test1)
```

```python
model_score_r1 = model_rf_smote.score(xr_test1, yr_test1)
```

```python
print(model_score_r1)
print(metrics.classification_report(yr_test1, yr_predict1))
```

```python
from xgboost import XGBClassifier
model_xg = XGBClassifier()
```

```python
model_xg.fit(xr_train1,yr_train1)
```

```python
y_xg = model_xg.predict(xr_test1)
```

```python
acc3 = accuracy_score(yr_test1, y_xg)
print(f"XGBoost Accuracy: {acc3}")
```
